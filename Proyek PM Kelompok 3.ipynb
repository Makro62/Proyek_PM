# Lower

```python
#proses membuat semua kalimat dalam comment mejadi lowecase(huruf kecil)
data['comment'] = data['comment'].str.lower()
```

```python
data['comment']
```

# Punctuation

```python
punctuation = string.punctuation
def remove_punctuation(text):
    return text.translate(str.maketrans('', '', punctuation))
```

Menghapus tanda baca

```python
data['comment'] = data['comment'].apply(lambda text: remove_punctuation(text))
```

```python
data['comment']
```

Menghapus angka

```python
data['comment'] = data['comment'].str.replace(r'[^a-zA-Z ]\s?',r'', regex=True)
```

```python
data['comment']
```

# Tokenization

```python
data['tokenized_comment'] = data.apply(lambda row: nltk.word_tokenize(row['comment']), axis=1)
```

```python
data['tokenized_comment']
```

Menghapus Stopwords

```python
from nltk.corpus import stopwords

list_stopwords = stopwords.words('indonesian')
list_stopwords = set(list_stopwords)
def stopwords_removal(words):
    return [word for word in words if not word in list_stopwords]
data['stopwords_comment'] = data['tokenized_comment'].apply(stopwords_removal) 
```

```python
data['stopwords_comment']
```

```python
normalizad_word = pd.read_csv("D://colloquial-indonesian-lexicon.csv") 
    

normalizad_word_dict = {}

for index, row in normalizad_word.iterrows():
    if row[0] not in normalizad_word_dict:
        normalizad_word_dict[row[0]] = row[1] 

def normalized_term(document):
    return [normalizad_word_dict[term] if term in normalizad_word_dict else term for term in document]

data['comment_normalized'] = data['stopwords_comment'].apply(normalized_term)

data['comment_normalized'].head(10)
```

# Mencocokan ke Sastrawi

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import swifter
factory = StemmerFactory()
stemmer = factory.create_stemmer()
def stemmed_wrapper(term):
    return stemmer.stem(term)

term_dict = {}

for document in data['comment_normalized']:
    for term in document:
        if term not in term_dict:
            term_dict[term] = ' '
            
print(len(term_dict))
print("------------------------")

for term in term_dict:
    term_dict[term] = stemmed_wrapper(term)
    print(term,":" ,term_dict[term])
    
print(term_dict)
print("------------------------")

def get_stemmed_term(document):
    return [term_dict[term] for term in document]
data['comment_stemmed'] = data['comment_normalized'].head(100).swifter.apply(get_stemmed_term)
print(data['comment_stemmed'])
```
